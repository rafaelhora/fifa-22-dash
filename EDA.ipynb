### Cleaning columns with Currency, height and weight

```python
df['Wage'] = df['Wage'].replace({'€': '', ',': '', 'K':'000', '\.':''}, regex=True).astype(float)
df['Wage']
```

```python
df[['Name', 'Wage']].sort_values('Wage',ascending=False).head(20)
```

```python
df['Release Clause'] = df['Release Clause'].replace({'€': '', ',': '', 'K':'000','M':'000000', '\.':''}, regex=True).astype(float)
df['Value'] = df['Value'].replace({'€': '', ',': '', 'K':'000', 'M':'000000', '\.':''}, regex=True).astype(float)
df['Weight'] = df['Weight'].replace({'kg':''}, regex=True).astype(float)
df['Height'] = df['Height'].replace({'cm':''}, regex=True).astype(float)
```

### Best Squad

```python
positions = df['Best Position'].unique()
positions
```

```python
best_player = pd.concat([pd.DataFrame(pd.DataFrame(df[df['Best Position'] == positions[i]]).sort_values('Best Overall Rating', ascending=False).iloc[0:1]) for i in range(len(positions))],
          ignore_index=True)
best_player
```

### Ratio of Wage and Vale vs Overall Rating

```python
ratings_ratio = df.groupby('Club').mean()[['Best Overall Rating','Value','Wage']].reset_index().sort_values('Best Overall Rating', ascending = False)

ratings_ratio['ratio_wage'] = ratings_ratio['Wage'] / ratings_ratio['Best Overall Rating']
ratings_ratio['ratio_value'] = ratings_ratio['Value'] / ratings_ratio['Best Overall Rating']
ratings_ratio
```

```python
ratings_ratio.sort_values('ratio_wage', ascending=False).head(10)
```

```python
ratings_ratio.sort_values('ratio_value', ascending=False).head(10)
```

### Player with the most potential

```python
promising_players = df[['ID', 'Name', 'Age','Club','Best Position','Best Overall Rating', 'Potential','Value', 'Wage']]
promising_players['future_growth'] = promising_players['Potential'] - promising_players['Best Overall Rating']
promising_players.sort_values(['future_growth', 'Potential'], ascending = False).head(20)
```
